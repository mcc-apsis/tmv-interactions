# Some recipes for analysing document data from the apsis platform

Using python on the APSIS server, we can interact directly with the database and easily traverse the relationships between the database's different tables

Here are a few examples plotting some basic scientometric data, and showing differences between relevant and not relevant documents

```python
# Load some libraries for reading the data from the database and plotting 

import django
import sys, os
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append('/home/galm/software/django/tmv/BasicBrowser/')
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "BasicBrowser.settings")
django.setup()

from scoping.models import *

from django.db.models import Count, Sum

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords as sw
import numpy as np
```

```python
# get the query we are interested in -- 
q = Query.objects.get(pk=5309)

# get the documents associated with the query
docs = Doc.objects.filter(query=q,docproject__project=q.project)

#how many documents do we have?
docs.count()
```

```python
# How many documents are relevant or not
rel = list(docs.filter(docproject__project=q.project).values('docproject__relevant').annotate(
    n = Count('pk') # Count the pk - primary keys - of documents with each value
).order_by('-n'))

for r in rel: # get the full name of relevance
    r['relevant'] = DocProject.Relevance[r['docproject__relevant']][1]

# Put this into a dataframe
rel_df = pd.DataFrame.from_dict(rel) # we have to wrap the queryset in a list, so pandas can read it properly
rel_df.tail() # what do the last few rows look like?

fig, ax = plt.subplots()

ax.bar(rel_df.index,rel_df['n'])
ax.set_xticks(rel_df.index)
ax.set_xticklabels(rel_df['relevant'])

plt.show()
```

```python
# how many documents in each year of each relevance category?
by_year = list(docs.values('PY','docproject__relevant').annotate(
    n = Count('pk') # Count the pk - primary keys - of documents in each value of PY (year)
).order_by('PY'))

for r in by_year: # get the full name of relevance
    r['relevant'] = DocProject.Relevance[r['docproject__relevant']][1]

# Put this into a dataframe
by_year_df = pd.DataFrame.from_dict(list(by_year)) # we have to wrap the queryset in a list, so pandas can read it properly
by_year_df.tail() # what do the last few rows look like?
```

```python

for name, group in by_year_df.groupby('relevant'):
    plt.scatter(group['PY'],group['n'],label=None,s=2)
    plt.plot(group['PY'],group['n'],label=name)
    # Add some labels
    plt.xlabel('Publication year')
    plt.ylabel('Number of documents')
    
plt.legend()
```

```python
# Or we can do bars next to each other
fig, ax = plt.subplots(figsize=(12,8))

offset_size = 0.28

offset = - offset_size
for name, group in by_year_df.groupby('relevant'):
    plt.bar(
        [x + offset for x in group['PY']],
        group['n'],
        width=offset_size,
        label=name,
        linewidth=0.6,
        edgecolor="grey"
    )
    offset+=offset_size
    
# Add some labels
plt.xlabel('Publication year')
plt.ylabel('Number of documents')
    
plt.legend()
```

```python
# What keywords have relevant/irrelevant documents got?
doc_kw = list(docs.values('kw__text','docproject__relevant').annotate(
    n = Count('pk')
).order_by('-n'))
for r in doc_kw: # get the full name of relevance
    r['relevant'] = DocProject.Relevance[r['docproject__relevant']][1]
doc_kw_df = pd.DataFrame.from_dict(doc_kw) 

## Uncomment the line below if we want to see nones in the graph
# doc_cat_df.fillna(value="None",inplace=True)

doc_kw_df.head()
```

```python
# here's a little function for plotting bar graphs of the frequency of a category 
# across relevant and not relevant documents
def plot_relevance_categories(df,cat_variable,n):
    cat_sets = []
    for i, (name, group) in enumerate(df.groupby('relevant')):
        group = group.sort_values('n',ascending=False).reset_index(drop=True).head(n)
        cats = group[cat_variable]
        cats = [x for x in cats if x is not None]
        cat_sets.append(cats)

    #set up 3 axes
    fig, axs = plt.subplots(3,figsize=(12,12))

    # Go through the sets of top keywords
    for i, cats in enumerate(cat_sets):

        ax = axs[i]
        ax.set_ylabel("Frequency")
        ax.set_xlabel("Keyword")

        cats = [x for x in cats if x is not None]
        # plot the frequency of these keywords in each group
        offset = -0.25
        for j, (name, group) in enumerate(df.groupby('relevant')):
            group = group[group[cat_variable].isin(cats)].reset_index(drop=True)
            group['cat'] = group[cat_variable].astype("category")
            group['cat'].cat.set_categories(cats,inplace=True)
            group = group.sort_values('cat').reset_index(drop=True)

            ax.bar(group.index + offset,group['n'],width=0.25, label=name)
            offset+=0.25

            if i==j:
                ax.set_xticks(group.index)
                ax.set_xticklabels(group['cat'],rotation = 30,ha="right")
                
    axs[2].legend()

    fig.tight_layout()
```

```python
# Get the top n keywords from each group
plot_relevance_categories(doc_kw_df,"kw__text",15)
```

```python
# Now we do the same for web of science categories
doc_cat = docs.values('wc__text','docproject__relevant').annotate(
    n = Count('pk')
).order_by('-n')
for r in doc_cat: # get the full name of relevance
    r['relevant'] = DocProject.Relevance[r['docproject__relevant']][1]
doc_cat_df = pd.DataFrame.from_dict(list(doc_cat)) 

## Uncomment the line below if we want to see nones
# doc_cat_df.fillna(value="None",inplace=True)

plot_relevance_categories(doc_cat_df,"wc__text",15)

```

### Now we can start doing some basic text mining


```python

stopwords = set(sw.words('english'))

worddocs = docs.filter(content__iregex='\w',docproject__relevant__in=[1,2]).order_by('id')

texts = list(worddocs.values_list('title',flat=True))
texts = [" ".join([t for t in [x.title,x.content,x.wosarticle.de] if t is not None]) for x in worddocs]
relevance = list(worddocs.order_by('id').values_list('docproject__relevant',flat=True))

vec = CountVectorizer(
    stop_words = stopwords,
    ngram_range = (1,2),
    min_df = 5
)
dtf = vec.fit_transform(texts)
print(f"found {dtf.shape[1]} words in {dtf.shape[0]} documents")

sum_words = dtf.sum(axis=0)
corpus_freq = np.diff(dtf.tocsc().indptr)

words_freq = [(word, corpus_freq[idx]) for word, idx in vec.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)

words_freq[:5]
```

```python
fig, axs = plt.subplots(2,figsize=(12,8))
for i,x in enumerate([(1,2),(2,1)]):
    t = f"Most common phrases never appearing in category: {DocProject.Relevance[x[0]][1]}"
    dtf_1 = dtf[np.argwhere(np.array(relevance)==x[0])[:,0],:]
    dtf_2 = dtf[np.argwhere(np.array(relevance)==x[1])[:,0],:]
    scores_1 = np.diff(dtf_1.tocsc().indptr)
    scores_2 = np.diff(dtf_2.tocsc().indptr)
    tids = np.argwhere(scores_1==0)[:,0]
    df = pd.DataFrame({
        "tids": tids,
        "freq": scores_2[scores_1==0],
        "term": np.array(vec.get_feature_names())[tids]
    })
    df = df.sort_values('freq',ascending=False).reset_index(drop=True).head(15)
    ax = axs[i]
    ax.bar(df.index,df['freq'])
    ax.set_xticks(df.index)
    ax.set_xticklabels(df['term'],rotation = 45,ha="right")    
    ax.set_title(t)

fig.tight_layout()
```

```python
from sklearn.naive_bayes import MultinomialNB, ComplementNB

clf = ComplementNB(alpha=0.1)
#clf = MultinomialNB(alpha=0.1)
clf.fit(dtf,relevance)

coefs = clf.feature_log_prob_
for i in range(coefs.shape[0]):
    print("######\n")
    top_ids = np.argsort(coefs[i,:])
    #for tid in top_ids[:5]:
    #    print(vec.get_feature_names()[tid])
    #    #print(np.exp(coefs[i,tid]))
    #    print(coefs[i,tid])
    for tid in top_ids[-10:]:
        print(vec.get_feature_names()[tid])
        #print(np.exp(coefs[i,tid]))
        print(coefs[i,tid])
```
